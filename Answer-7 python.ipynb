```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data from Question 4 into a DataFrame
data = pd.read_csv('meteorite_data.csv')

# Convert the 'year' column to datetime format for easier manipulation
data['Year'] = pd.to_datetime(data['Year'], errors='coerce')

# Get all the Earth meteorites that fell before the year 2000
earth_meteorites_before_2000 = data[data['Year'] < pd.to_datetime('2000-01-01')]

# Get all the Earth meteorites' coordinates that fell before the year 1970
earth_meteorites_before_1970 = data[(data['Year'] < pd.to_datetime('1970-01-01')) & (data['Latitude'].notna()) & (data['Longitude'].notna())]

# Assuming the mass of meteorites is in kg, get all meteorites with a mass greater than 10000 kg
meteorites_mass_greater_than_10000kg = data[data['Mass (g)'] > 10000]

# Plotting the data

# Plotting the distribution of meteorite falls over the years
plt.figure(figsize=(12, 6))
data['Year'].dt.year.value_counts().sort_index().plot(kind='bar')
plt.xlabel('Year')
plt.ylabel('Number of Meteorite Falls')
plt.title('Distribution of Meteorite Falls over the Years')
plt.show()

# Plotting the locations of meteorite falls on a map
plt.figure(figsize=(10, 8))
plt.scatter(data['Longitude'], data['Latitude'], marker='o', alpha=0.5)
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Locations of Meteorite Falls')
plt.show()

```
